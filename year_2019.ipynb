# Crime Data 2019

```python
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
#Import csv
raw2019 = pd.read_csv('/csv/2019-PART_I_AND_II_CRIMES.csv', sep = ',')
```

```python
#Rows and columns
raw2019.shape
```

```python
#Data Types
raw2019.dtypes
```

```python
#Convert to datetime
raw2019['INCIDENT_DATE'] = pd.to_datetime(raw2019['INCIDENT_DATE'], format = '%m/%d/%Y %I:%M:%S %p')
raw2019['INCIDENT_REPORTED_DATE'] = pd.to_datetime(raw2019['INCIDENT_REPORTED_DATE'], format = '%m/%d/%Y')

#Sort by Incident_Date, Incident_Reported_Date, ID
raw2019 = raw2019.sort_values(by=['INCIDENT_DATE', 'INCIDENT_REPORTED_DATE', 'INCIDENT_ID'], ascending=[True, True, True])
raw2019.dtypes
```

```python
#Filter by Q1/Q2 2019
start_date = '2018-12-31 23:59:59'
end_date = '2019-06-30 23:59:59'

mask = (raw2019['INCIDENT_DATE'] > start_date) & (raw2019['INCIDENT_DATE'] <= end_date)
s2019 = raw2019.loc[mask]
s2019.reset_index(drop=True)
```

```python
#Filter columns
f2019 = s2019[["INCIDENT_DATE", "CATEGORY", "STAT", "STAT_DESC", "ADDRESS", "STREET", "CITY", "ZIP", "INCIDENT_ID", "REPORTING_DISTRICT", "GANG_RELATED", "UNIT_ID", "UNIT_NAME", "LONGITUDE", "LATITUDE", "PART_CATEGORY"]]
f2019.reset_index(drop=True)
```

```python
#Count null values for each column
f2019.isnull().sum()
```

```python
#Change data types
f2019 = pd.DataFrame(f2019)
f2019[['CATEGORY', 'STAT_DESC', 'ADDRESS', 'STREET', 'CITY', 'INCIDENT_ID', 'GANG_RELATED', 'UNIT_ID', 'UNIT_NAME']] = f2019[['CATEGORY', 'STAT_DESC', 'ADDRESS', 'STREET', 'CITY', 'INCIDENT_ID', 'GANG_RELATED', 'UNIT_ID', 'UNIT_NAME']].astype('string')
#f2019['ZIP'] = f2019['ZIP'].astype('Int64')
f2019.dtypes
```

```python
import geopy

dfT = f2019.head(10)

def get_zipcode(df, geolocator, lat_field, lon_field):
    if ((dfT[lat_field].notna()) & (dfT[lon_field].notna())):
        location = geolocator.reverse((df[lat_field], df[lon_field]))
        return location.raw['address']['postcode']
    else:
        return 0

geolocator = geopy.Nominatim(user_agent='1234')

#zCode = f2019.apply(get_zipcode, geolocator=geolocator, axis=1, lat_field='LATITUDE', lon_field='LONGITUDE')
zmask = (dfT['ZIP'].isnull() and (dfT['LATITUDE'].notna()))
dfT.loc[zmask, 'ZIP'] = dfT.apply(get_zipcode, geolocator=geolocator, axis=1, lat_field='LATITUDE', lon_field='LONGITUDE')
dfT
#zCode
```
