```python
import pandas as pd
```

```python
zip = pd.read_csv('C:/Users/hanna/Documents/CSV/zip_to_lat_lon_North America.csv', sep = ',')
mask = (zip['state'] == 'California')
caZ = zip.loc[mask]
caZ["postalcode"] = caZ["postalcode"].astype(dtype = 'Int64')
caZ.reset_index(drop=True)
```

```python
caZ = caZ[['postalcode', 'place', 'statecode', 'province_or_county','latitude', 'longitude']]
caZ.reset_index(drop=True)
```

```python
#Export to csv
caZ.to_csv('C:/Users/hanna/Documents/CSV/CA_Zip_Codes.csv')
```
